## Year

```python
df["Year"] = np.nan
```

```python
for i in range(0,len(df)):
    text = df.loc[i,"Date"]
    numbers = [int(n) for n in text.split() if n.isdigit()]
    for d in numbers:
        d = str(d)
        if len(d) == 4 or len(d) == 3 and d.isdigit():
            df["Year"].loc[i] = d
```

```python
df["Year"].loc[df["Date"]=='60 BC'] = 60
df["Year"].loc[df["Date"]=='17 AD'] = 17
df["Year"].loc[df["Date"]=='February 5, 62 AD'] = 62
```

```python
df['Year'] = df['Year'].astype('int')
#df['Year'] = df['Year'].astype(np.float).astype("Int32")
```

```python
for i in range(0, len(df)):
    if "BC" in df.loc[i,"Date"]:
        df.loc[i,"Year"] = -df.loc[i,"Year"] 
```

## Magnitude

```python
df["Magnitude"].unique()
```

```python
df["Mag"] = df["Magnitude"]
df["Mag"] = df["Mag"].str.replace("[\(\[].*?[\)\]]","")
```

```python
for i in range(0,len(df)):
    try:
        text = df.loc[i,"Mag"]
        numbers = re.findall(r"[-+]?\d*\.\d+|\d+", str(text))
        numbers_f = list(map(float, numbers))
        num = max(numbers_f)
        df["Mag"].loc[i] = num
    except:
        df["Mag"].loc[i] = np.nan

```
